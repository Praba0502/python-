```python
import pandas as pd

# Input data
item = {'Company': ['Google', 'Google', 'BMW', 'BMW', 'Mrf', 'Mrf'],
        'Person': ['ram', 'raj', 'ramu', 'john', 'cena', 'rock'],
        'Sales': [200, 234, 345, 654, 789, 419]}

# Creating the DataFrame
Dove = pd.DataFrame(item)

# Group by 'Company' and calculate mean only on the 'Sales' column (numeric data)
result = Dove.groupby('Company')['Sales'].mean()

# Display the result
print(result)
```

```python
Dove.groupby('Company').count()
```

```python
Dove.groupby('Company').max()
```

```python
Dove.describe()
```

```python
Dove.std
```

```python
Dove.groupby('Company')['Sales'].median()
```

Merging, Joining and Concatenating

Namma kitta 3 ways irukku Dataframe combine pantrathukku: Merging, Joining, and Concatenating.

```python
left=pd.DataFrame({'Key':['k0','k1','k2','k3'],
                   'A':['a0','a1','a2','a3'],
                   'B':['b0','b1','b2','b3']})
```

```python
right=pd.DataFrame({'Key':['k0','k1','k2','k3'],
                    'C':['c0','c1','c2','c3'],
                    'D':['d0','d1','d2','d3']})
```

```python
left
```

```python
right
```

```python
pd.merge(right,left,how='inner',on='Key')
```

```python
pd.merge(left,right,how='outer',on='Key')
```

```python
left=pd.DataFrame({'Key1':['k0','k1','k2','k3'],
                   'Key2':['k0','k1','k0','k1'],
                   'A':['a0','a1','a2','a3'],
                   'B':['b0','b1','b2','b3']})
```

```python
right=pd.DataFrame({'Key1':['k0','k1','k2','k3'],
                    'Key2':['k0','k0','k0','k0'],
                    'C':['c0','c1','c2','c3'],
                    'D':['d0','d1','d2','d3']})
```

```python
pd.merge(left,right,on=['Key1','Key2'])
```

```python
print(left)
print(right)
```

```python
pd.merge(left,right,how='outer',on=['Key1','Key2'])
```

```python
pd.merge(left,right,how='inner',on=['Key1','Key2'])
```

```python
pd.merge(left,right,how='outer',on=['Key1','Key2'])
```
